```python
##datatypes in csv
data = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11')
```

```python
df=pd.read_csv(StringIO(data),dtype={'a':int,'b':float,'c':int})
```

```python
df.info()
```

```python
data = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data)) #we want to make the previous index column in new data set
```

```python
pd.read_csv(StringIO(data),index_col=0)
```

```python
pd.read_csv(StringIO(data),index_col=0,usecols=['index','b','c'])
```
